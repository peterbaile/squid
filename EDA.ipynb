```
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```
# Readers over time (total readers by month)

data = pd.read_csv('drive/MyDrive/CIS520 Project/data set/content_views_merged.csv')
```

```
# Histogram of articles and views
```

```
data['published_date'] = data['published_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
data['year_month'] = data['published_date'].apply(lambda x: datetime.strftime(x, '%Y-%m'))
data['year_month'] = data['year_month'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
```

```
data.head()
```

```
data.groupby('year_month')['pageviews'].count().plot.line(xlabel = 'Year', ylabel = 'Numbe of Articles', title = 'Monthly Articles over Time')
```

```
data.groupby('year_month')['pageviews'].mean().plot.line(xlabel = 'Year', ylabel = '3 day Average Views', title = 'Average 3-days Views per Article over Time')
```
